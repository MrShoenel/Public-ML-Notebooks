```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def alpha(err):
    return np.log((1-err)/err)
def adjust(err):
    return (1-err)/err
```

```python
errs = np.arange(start=0.001, stop=1, step=0.001)  # an array from 0 to 1 with a step of 0.01
alphas = alpha(errs)
plt.plot(errs, alphas)
plt.ylabel("log((1-err)/err)")
plt.xlabel("err")
```

```python
adjusts = adjust(errs)
plt.plot(errs, adjusts)
plt.ylabel("(1-err)/err")
plt.xlabel("err")
```
