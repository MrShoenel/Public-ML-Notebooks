```matlab
f = @(x) -14.6729+4.8266*x-0.4855*x.^2+0.0312*x.^3-0.0055*x.^4+0.0009*x.^5;
x = -10:0.1:15;
y = f(x);
plot(x,y)
```

```matlab
a = -100;
b = 100;
r1 = (b-a).*rand(1,10+1+15) + a;
r2 = (b-a).*rand(1,10+1+15) + a;
xx = -10:1:15;
s1= r1+f(xx);
s2= r2+f(xx);
plot(x,y)
hold on
scatter(xx,s1)
scatter(xx,s2,"*","g")
```

```matlab
lm1 = fitlm(xx,s1);
beta1 = lm1.Coefficients.Estimate;
lf = @(x,beta) beta(1)+x.*beta(2);
plot(x,y)
hold on
scatter(xx,s1)
plot(x,lf(x,beta1),"r")
```

```matlab
lm2 = fitlm(xx,s2);
beta2 = lm2.Coefficients.Estimate;
plot(x,y)
hold on
scatter(xx,s2,"*","g")
plot(x,lf(x,beta2),"g")
```

```matlab
%%file eval_polynomial.m

function y = eval_polynomial(x,beta)
    n = length(beta);
    y = 0;
    for i=1:n
        y = y + x.^(n-i) * beta(i);
    end
end
```

```matlab
pm1 = polyfit(xx,s1,10);
pf = @(x,beta) eval_polynomial(x,beta);
plot(x,y)
hold on
scatter(xx,s1)
plot(x,pf(x,pm1),"r")
```

```matlab
pm2 = polyfit(xx,s2,10);
plot(x,y)
hold on
scatter(xx,s2,"*","g")
plot(x,pf(x,pm2),"g")
```
